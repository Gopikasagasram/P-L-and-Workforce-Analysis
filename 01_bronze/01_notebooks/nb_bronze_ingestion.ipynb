```python
%sql
use schema bronze;
```

```python
accounts_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/accounts.csv", header=True)
company_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/company.csv", header=True)
departments_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/departments.csv", header=True)
employee_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/employee.csv", header=True)
general_ledgers_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/general ledgers.csv", header=True)
payroll_df= spark.read.csv("s3://p-l-workforce-analysis-dataset/raw data/payroll.csv", header=True)

```

```python
accounts_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.accounts")
```

```python
company_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.company")
employee_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.employees")
departments_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.departments")
general_ledgers_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.general_ledgers")
payroll_df.write.mode("overwrite").saveAsTable("pl_workforce_analysis.bronze.payroll")
```
